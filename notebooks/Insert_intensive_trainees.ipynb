```python
def create_batch (batch_params: dict):
  query = """mutation createBatch(
  $batch: Int
  $class_link: String
  $communication_link: String
  $additional_info: JSON
) {
  createBatch(data: {
    Batch:$batch
    Class_link:$class_link
    Communication_link:$communication_link
    additional_info:$additional_info
    
  }) {
    data {
      id
    }
  }
}"""
  res_json  =  tip.sg.Select_from_table(query=query, variables={"batch": batch_params['batch'],
                                                              "class_link": batch_params['class_link'],
                                                              "communication_link": batch_params['communication_link'],
                                                              "additional_info": batch_params['additional_info']})
  return res_json
```

```python
def parse_trainee_info(res_json):
    trainee_info = []
    for i in res_json['data']['usersPermissionsUsers']['data']:
        params = {
            'userId': i['id'],
            'email': i['attributes']['email'],
            'name': i['attributes']['username']
        }
        trainee_info.append(params)
    df = pd.DataFrame(trainee_info)
    df['batch']= intensive_batch9_trainee_config.batch
    df['role']= "trainee"
    return df
```
