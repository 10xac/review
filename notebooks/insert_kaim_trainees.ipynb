```python
def parse_trainee_info(res_json):
    trainee_info = []
    for i in res_json['data']['usersPermissionsUsers']['data']:
        params = {
            'userId': i['id'],
            'email': i['attributes']['email'],
            'name': i['attributes']['username']
        }
        trainee_info.append(params)
    df = pd.DataFrame(trainee_info)
    df['batch']= kiam_batch2_trainee_config.batch
    df['role']= "trainee"
    return df
```
