```python
import pandas as pd
def parse_trainee_info(res_json):
    trainee_info = []
    for i in res_json['data']['usersPermissionsUsers']['data']:
        params = {
            'userId': i['id'],
            'email': i['attributes']['email'],
            'name': i['attributes']['username']
        }
        trainee_info.append(params)
    df = pd.DataFrame(trainee_info)
    df['batch']= u2j_batch3_trainee_config.batch
    df['role']= "trainee"
    return df
```

```python
### to update existing group
def get_all_user_without_group ():
    query = """query getallUserID{
    allUsers(
        pagination:{start:0,limit:2000}
        filters:{groups:{id:{eq:null}}}){
        meta{
        pagination{
            total
        }
        }
        data{
        id
        }
    }
    }"""
    result_json = tip.sg.Select_from_table(query=query, variables=None)
    without_group = [str(i['id']) for i in result_json['data']['allUsers']['data']]
    return without_group
```

```python
without_group = get_all_user_without_group ()
len( without_group)
```

```python
#get_allUser_by_groupId(groupId)
def get_allUser_by_groupId( groupId):
    query = """query getallUserID($groupID:ID){
      allUsers(
        pagination:{start:0,limit:2000}
        filters:{groups:{id:{eq:$groupID}}}){
        meta{
          pagination{
            total
          }
        }
        data{
          id
        }
      }
    }"""
    resu_json = tip.sg.Select_from_table(query=query, variables={"groupID":groupId})
    with_group = [str(i['id']) for i in resu_json['data']['allUsers']['data']]
    return with_group
```
