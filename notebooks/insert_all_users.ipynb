```python
def get_alluser_ids_based_on_list_of_emails(list_of_emails):
    query = """query alluserdata($emails:[String]){
            allUsers(
                pagination:{start:0,limit:100}
                filters:{email:{in:$emails}}){
                data{
                id
                attributes{
                    email
                    
                }
                }
            }
            }"""
    result_json = iar.sg.Select_from_table(query=query, variables={"emails":list_of_emails})
    return result_json['data']['allUsers']['data']
```

```python
def update_allusers_with_previous_batch_applicant(ids_to_change, current_batch):
    for i in ids_to_change:
        query = """mutation updateAlluserBatch($id:ID!, $batch:Int){    
            updateAllUser(id:$id,data:{Batch:$batch}){
            data{
                id
            }
            }
        }"""

        resjson = iar.sg.Select_from_table(query=query, variables={"id":i, "batch":current_batch})
        print(resjson)
```
